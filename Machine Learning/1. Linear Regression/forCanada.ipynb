```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from sklearn import linear_model
style.use('dark_background')
```

```python
%matplotlib inline
plt.scatter(df.year, df.percapitaincome)
```
