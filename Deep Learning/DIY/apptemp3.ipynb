```python
import pandas as pd
df = pd.read_csv("Salary_data.csv")
df.head()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
dummies = pd.get_dummies(df[['Gender', 'Education Level', 'Job Title']])
dummies.head()
```

```python
df = df.drop(['Gender', 'Education Level', 'Job Title'], axis='columns')
df = pd.concat([dummies, df], axis='columns')
df.head()
```

```python
X = df.drop(['Salary'], axis='columns')
y = df.Salary
y.head()
```

```python
def Standardize(df):
    for feature_name in df.columns: # The train data
        if df[feature_name].dtype in ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']:
            mean_value = df[feature_name].mean()
            std_value = df[feature_name].std()
            df[feature_name] = (df[feature_name] - mean_value) / std_value
        elif df[feature_name].dtype == 'bool':
            df[feature_name] = df[feature_name].astype(float)
    return df
```
